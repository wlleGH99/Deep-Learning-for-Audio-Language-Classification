```python
# for Taoteba files, which were all mp3 snippets
import os
from os import path
from pydub import AudioSegment

directory = 'Spanish/'
for filename in os.listdir():
    #file = os.path.join(directory, filename)
    if filename.endswith(".mp3"):
      print(filename)
      sound = AudioSegment.from_file(filename)
      sound.export(directory + filename[:-4] + ".wav", format="wav")
```
